<h2>Indexing</h2>

```python
flights['Date'] = flights['Date'].astype(str)
flights['Date'] = pd.to_datetime(flights['Date'],format="%Y%m")
flights = flights.set_index('Date')
```

```python
flights.head()
```

```python
flights.index
```

```python
print(flights.index.names, flights.index.values)
```

```python
flights = flights.to_period(freq='M')
```

```python
type(flights.index.values[0])
```

```python
flights.head()
```

```python
flights.loc['2005-02'].head()
```

```python
flights.set_index('Origin',append=True).head()
```

```python
ORD = flights[flights['Origin'] == 'ORD']
SFO = flights[flights['Origin'] == 'SFO']
```

```python
ORD.head()
```

```python
SFO.head()
```

```python
SFO.groupby(level=0)
```

```python
SFO = SFO.groupby(level=0).sum()
ORD = ORD.groupby(level=0).sum()
SFO.head()
```

```python
ORD.head()
```

```python
(ORD + SFO).head()
```

```python
APC = flights[flights['Origin'] == 'APC']
APC = APC.groupby(level=0).sum()
APC
```

```python
SFO.head()
```

```python
(SFO + APC).head()
```

```python
SFO.add(APC,fill_value=0).head()
```

<h2>Plotting</h2>

```python
flights.plot(y = 'Distance',kind='hist',bins=20,weights=flights['Flights'])
plt.ylabel('Flights')
plt.show()
```

```python
months = flights[['Flights','Passengers']].groupby(level=0).sum()
```

```python
months.head()
```

```python
months.plot(y='Passengers')
plt.show()
```

```python
months.plot(layout=(2,1),subplots=True)
plt.show()
```

```python
months.plot(secondary_y='Flights')
plt.show()
```
