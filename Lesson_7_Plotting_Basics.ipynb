## Matplotlib

So we have all seen plotting in R, matlab, excel, etc.  Matplotlib is the most extensive python package for plotting, created by the late John Hunter, and it is compatible with Numpy and Pandas.  It has extensive 2D plotting functions, some 3D plotting, and extensions for geographical data (which we will go over later in the class).  The common 2D plots that we commonly make are `plot, scatter, hist, bar, pie, boxplot, contours`.  You can also draw basic objects on the plot, like lines, polygons, etc. but we won't be going over this in this lesson.  

Matplotlib provides all of the objects needed to plot and customize complicated figures, including adding text, polygons, axes, lines, etc.  It is very low level in the sense that you have very fine control over the inner workings of the plot itself.  This is in contrast to a tool like ggplot, which is very high level, and can provide quick, beautiful graphics by assuming that you will be performing certain operations.  We will talk about ggplot later in the course, but first, we should introduce matplotlib because it can do pretty much everything when it comes to static plotting.

Pyplot emulates matlab's plotting capabilities.  While matplotlib provides all of the objects that are used to build a plot, pyplot provides a simpler interface that uses matplotlib objects behind the scenes.  For example, the object `matplotlib.axes.Axes` is what we think of as the plot - which is where all of the lines, points, etc. are drawn.  We can remain ignorant of this object and its inner workings by using Pyplot which will create an instance of this object when we call `pyplot.figure()` (in fact the return value of this function is a figure and axes instance).  Consider the following code...  

```python
import numpy as np
import matplotlib.pyplot as plt

x = np.linspace(0, 1, 500)
y = np.sin(4 * np.pi * x) * np.exp(-5 * x)

plt.plot(x, y, zorder=10)
plt.show()
```

In order to do something more customized you may need to act on the `Axes` instances directly.  For example, the `fill` and `grid` methods are ways to add fill to the lines above and gridlines.  This example is can be found [here](https://matplotlib.org/examples/lines_bars_and_markers/fill_demo.html).

```python
fig, ax = plt.subplots()

ax.fill(x, y, zorder=10)
ax.grid(True, zorder=5)
plt.show()
```

The basic objects in matplotlib are,
- Figure: this contains everything that gets plotted.  It has children axes, which can be subfigures, the title, and the details of how the plot is actually printed.
- Axes: this is where the data lives, and it contains things like text, polygons, axis, lines, etc.
- Axis: this is actually the axis - how the scale is shown 
- Artists: everything is an artist, so this is sortof a grab-bag category.  Sometimes you construct artists when you are drawing very custom objects.

**Checkpoint:** Look through the [list of methods](http://matplotlib.org/api/pyplot_summary.html) in pyplot to see what is available.

## Images and style

Jupyter anticipated the use of matplotlib, so it has inline plotting.  You can even avoid the use of the `plt.show()` command with the `%matplotlib` magic command.

```python
%matplotlib inline
```

```python
fig, ax = plt.subplots()
ax.fill(x, y, zorder=10)
ax.grid(True, zorder=5)
```

I will typically begin my notebooks with this line and importing pyplot as plt.  You can also choose the style of your plot with `plt.style.use` such as the ggplot style.  You can see more details about customizing your style [here](https://matplotlib.org/users/style_sheets.html).

```python
plt.style.use('ggplot')
fig, ax = plt.subplots()
ax.fill(x, y, zorder=10)
ax.grid(True, zorder=5)
```

Finally, you can use `imshow` to plot images in python.  Images can be thought of as 2-dimensional arrays of pixel values.  If the plot is in color then each pixel value will consist of an RGB value.  So it is really an $n \times m \times 3$ array for a color n by m pixel image.  The nice thing about this is that you can read a png file into an array using `imread`, manipulate it using numpy, and then show it with `imshow`.  In the event that you only have a single pixel value (such as a grayscale image) and you want to view it in color then you will need a colormap.  A colormap will convert your single pixel value into RGB values.

**Checkpoint:** Read the image tutorial [here](https://matplotlib.org/users/image_tutorial.html).
